```python
# MEMBUAT LINE CHART

import csv
import matplotlib.pyplot as plt

# Membaca data dari file CSV
data = []
with open('UAS-Python/{}.csv'.format(key), 'r') as file:
    reader = csv.reader(file)
    next(reader)  # Mengabaikan header
    for row in reader:
        data.append(row)

# Mengambil kolom harga
harga = [float(row[1]) for row in data]

# Mengatur posisi x pada sumbu x
posisi_x = range(1, len(harga) + 1)

# Membuat line chart
plt.plot(posisi_x, harga)
plt.xlabel('Nomor Produk')
plt.ylabel('Harga')
plt.title('Line Chart Harga Produk')
plt.show()
```

```python
# MEMBUAT BAR CHART

import csv
import matplotlib.pyplot as plt

data = []  # List untuk menyimpan data dari file CSV
addresses = []  # List untuk menyimpan alamat toko
prices = []  # List untuk menyimpan harga produk

# Membaca data dari file CSV
with open('UAS-PYTHON/{}.csv'.format(key), 'r') as file:
    reader = csv.reader(file)
    next(reader)  # Mengabaikan header
    for row in reader:
        data.append(row)
        addresses.append(row[2])
        prices.append(float(row[1]))

# Membuat bar chart
plt.bar(addresses, prices)
plt.xlabel('Alamat')
plt.ylabel('Harga')
plt.title('Bar Chart Harga Produk Berdasarkan Alamat')
plt.xticks(rotation=90)  # Memutar label alamat untuk menghindari tumpang tindih
plt.tight_layout()  # Mengatur tata letak agar tidak tumpang tindih
plt.show()
```

```python
# MEMBUAT HISTOGRAM

import csv
import matplotlib.pyplot as plt

addresses = []  # List untuk menyimpan data alamat
prices = []  # List untuk menyimpan harga produk

# Membaca data dari file CSV
with open('UAS-Python/{}.csv'.format(key), 'r') as file:
    reader = csv.reader(file)
    next(reader)  # Mengabaikan header
    for row in reader:
        addresses.append(row[2])  # Mengambil alamat dari kolom ke-3 (indeks 2)
        prices.append(float(row[1]))  # Mengambil harga dari kolom ke-2 (indeks 1)

# Membuat histogram alamat
plt.figure(figsize=(10, 6))  # Mengatur ukuran plot
plt.hist(addresses, bins=10, edgecolor='black')  # Membuat histogram dengan 10 bins
plt.xlabel('Alamat')
plt.ylabel('Frekuensi')
plt.title('Histogram Alamat')
plt.xticks(rotation=45)  # Memutar label alamat agar tidak tumpang tindih
plt.tight_layout()  # Mengatur tata letak agar tidak tumpang tindih
plt.show()

# Membuat histogram harga
plt.figure(figsize=(10, 6))  # Mengatur ukuran plot
plt.hist(prices, bins=10, edgecolor='black')  # Membuat histogram dengan 10 bins
plt.xlabel('Harga')
plt.ylabel('Frekuensi')
plt.title('Histogram Harga')
plt.tight_layout()  # Mengatur tata letak agar tidak tumpang tindih
plt.show()
```

```python
# MEMBUAT DIAGRAM LINGKARAN

import csv
import matplotlib.pyplot as plt

data = []  # List untuk menyimpan data dari file CSV
addresses = []  # List untuk menyimpan alamat toko
prices = []  # List untuk menyimpan harga produk

# Membaca data dari file CSV
with open('UAS-Python/{}.csv'.format(key), 'r') as file:
    reader = csv.reader(file)
    next(reader)  # Mengabaikan header
    for row in reader:
        data.append(row)
        addresses.append(row[2])
        prices.append(float(row[1]))

# Menghitung total harga untuk setiap alamat
total_prices = {}
for i in range(len(addresses)):
    if addresses[i] in total_prices:
        total_prices[addresses[i]] += prices[i]
    else:
        total_prices[addresses[i]] = prices[i]

# Membuat diagram lingkaran
plt.pie(list(total_prices.values()), labels=list(total_prices.keys()), autopct='%1.1f%%')
plt.axis('equal')  # Mengatur sumbu x dan y menjadi sama panjang untuk mendapatkan lingkaran
plt.title('Diagram Lingkaran Harga Produk Berdasarkan Alamat')
plt.tight_layout()  # Mengatur tata letak agar tidak tumpang tindih
plt.show()
```
